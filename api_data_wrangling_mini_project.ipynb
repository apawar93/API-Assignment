These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
# Collecting data for all of 2017 for AFX_X
AFX_2017 = requests.get('https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&api_key=9jLzdSLspZwxM1-MPqxc')

```

```python
print(AFX_2017.json())
```

```python
# Converting JSON to Python dictionary

AFX_2017_Dict = AFX_2017.json()
print(AFX_2017_Dict)
type(AFX_2017_Dict)
```

```python
AFX_2017_Dict.keys()
```

```python
print(AFX_2017_Dict['dataset']['data'])
```

```python
AFX_2017_Data = AFX_2017_Dict['dataset']['data']
print(AFX_2017_Data)
type(AFX_2017_Data)
```

```python
AFX_2017_Opening_Price_Dict = {x[0]:x[1] for x in AFX_2017_Data}
print(AFX_2017_Opening_Price_Dict)
```

```python
AFX_2017_Opening_Price = list(AFX_2017_Opening_Price_Dict.values())
print(AFX_2017_Opening_Price)
```

```python
AFX_2017_Opening_Price = list(filter(None, AFX_2017_Opening_Price))
print(AFX_2017_Opening_Price)
```

```python
# Highest Opening Price
print("The highest opening price for AFX in 2017 was: ", max(AFX_2017_Opening_Price))
print("The lowest opening price for AFX in 2017 was: ", min(AFX_2017_Opening_Price))
```

```python
AFX_2017_high_low_Dict = {x[0]:x[1:3] for x in AFX_2017_Data}
print(AFX_2017_high_low_Dict)
```

```python
AFX_2017_High_Low = list(AFX_2017_high_low_Dict.values())
print(AFX_2017_High_Low)
```

```python
AFX_2017_High_Low = list(filter(None, AFX_2017_High_Low))
print(AFX_2017_High_Low)
```

```python
AFX_2017_HLDiff = [x[1]-x[0] if x[0] != None else 0 for x in AFX_2017_High_Low]
print(AFX_2017_HLDiff)
```

```python
# Highest change based on High and Low Prices for AFX
print("The highest change in price in 2017 for AFX was: ", max(AFX_2017_HLDiff))
```

```python
AFX_2017_traded_vol_Dict = {x[0]:x[6] for x in AFX_2017_Data}
print(AFX_2017_traded_vol_Dict)
```

```python
AFX_2017_traded_vol = list(AFX_2017_traded_vol_Dict.values())
print(AFX_2017_traded_vol)
```

```python
# Average traded in 2017 for AFX
print("The average traded volume for AFX in 2017 was: ", statistics.mean(AFX_2017_traded_vol))
```

```python
# Creating dictionary with daily price differences
AFX_2017_Daily_Diff = requests.get('https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X.json?start_date=2017-01-01&end_date=2017-12-31&transform=rdiff&api_key=9jLzdSLspZwxM1-MPqxc')

AFX_2017_Diff_Dict = AFX_2017_Daily_Diff.json()

print(AFX_2017_Diff_Dict)

type(AFX_2017_Diff_Dict)
```

```python
AFX_2017_Diff_Dict.keys()
```

```python
AFX_2017_Data_Diff = AFX_2017_Diff_Dict['dataset']['data']
print(AFX_2017_Data_Diff)
type(AFX_2017_Data_Diff)
```

```python
AFX_2017_Close_Diff_Dict = {x[0]:x[4] for x in AFX_2017_Data_Diff}
print(AFX_2017_Close_Diff_Dict)
```

```python
AFX_2017_Diff = list(AFX_2017_Close_Diff_Dict.values())
print(AFX_2017_Diff)
```

```python
print("The largest change in price between two days for AFX in 2017 was: ", max(AFX_2017_Diff))
```
